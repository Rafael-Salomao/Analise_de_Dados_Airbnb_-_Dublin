### Identificar outliers

Com base no que foi mencionado acima, iremos analisar a presença de outliers.

Em estatística, o outlier, que também pode ser chamado de valor atípico, se trata de uma observação de um valor de entrada que apresenta um grande afastamento no conjunto de valores em questão.

A existência de outliers implica, na grande parte dos casos, tendências indesejáveis e prejuízos na interpretação dos resutados. 

É preciso tratar os outliers devidamente a fim de reduzir os desequilíbrios por eles gerados na série. Para isso, iremos identificar os outliers de duas diferentes maneiras:

- Análise estatística das variáveis a partir do `.describe()`

- Plotar um gráfico `boxplots`, para uma análise mais visual.

Como vimos anteriormente, no plot dos `histogramas`, algumas variáveis apresentaram uma tendência há possuirem valores, possivelmente, discrepantes. No entanto, no presente projeto iremos restringir as análises nas variáveis `price` e `minimum_nights`.

Com a visualização dos valores estatísticos é possível inferir certas premissas, como:

- Para a variável `price`, 75% das entradas são menores que 150 sendo que o maior valor é de 1.173.721.

- Já para a variável `minimum_nights`, o valor máximo encontrado é de 1125 dias - equivalente a mais de dois anos de tempo mínimo de estadia.

- Para ambas variáveis:
 - Ao comparar os valores das medianas com as médias das entradas, há um forte indício dos maiores valores serem outliers.
 - Junto a isso, os desvios padrões são, relativamente, elevados em consideração as médias das entradas.

```
# mostrar a descrição das entradas numéricas utilizando a função .describe()
df[['price','minimum_nights']].describe()
```

```
# identificando os outliers para a variável `minimum_night`
q1_minimum_nights = df.minimum_nights.quantile(.25)
q3_minimum_nights = df.minimum_nights.quantile(.75)
IQR = q3_minimum_nights - q1_minimum_nights
print('O IQR da variável `minimum_nights` é: {}'.format(IQR))

# definindo limites
sup_minimum_nights = q3_minimum_nights + 1.5*IQR
inf_minimum_nights = q1_minimum_nights - 1.5*IQR
print('Limite superior para a variável `minimum_nights` é: {}'.format(sup_minimum_nights))
print('Limite superior para a variável `minimum_nights` é: {}'.format(inf_minimum_nights))
```

```
# identificando os outliers para a variável `price`
q1_price = df.price.quantile(.25)
q3_price = df.price.quantile(.75)
IQR = q3_price - q1_price
print('O IQR da variável `price` é: {}'.format(IQR))

# definindo limites
sup_price = q3_price + 1.5*IQR
inf_price = q1_price - 1.5*IQR
print('Limite superior para a variável `price` é: {}'.format(sup_price))
print('Limite superior para a variável `price` é: {}'.format(inf_price))
```

### Plot Boxplot

#### `minimum_nights`

A partir da visualização do gráfico de boxplot plotado utilizando os valores originais da variável `minimim_nights`, é possível notar a grandes disparidade seus valores. A presença dos outliers certamente irão prejudicar as futuras análises.

Para reduzir a influência dos valores dispersos, tomou-se como premissa o limite superior encontrado, no passo anterior, para adequar e reestruturar os valores da variável. A fim de avaliar realizar, posteriormente, a alteração e exclusão dos valores discrepantes, vê-se que tais entradas representam pouco mais de 12% do dataset total.

Em certas modelos de aprendizado de máquina, tal quantidade pode prejudicar de forma significativa o dataset. No entanto, para o intuito deste projeto, faz-se condizente a retirada desses elementos. A qualidade das análises terão uma melhora considerável, o que justifica o trade off desta decisão.

```
# minimum_nights
fig, ax = plt.subplots()
df.minimum_nights.plot(kind='box', vert = False, figsize=(12,3), ax=ax)
ax.set_title('\nBoxplot `minimum_nights`\n')
plt.show()

# ver quantidade de valores acima do limite superior encontrado anteriormente para minimum_nights
print('\nO dataset possui {} entradas'.format(df.shape[0]))
print('Para a variável `minimum_nights` conta com {} entradas acima de {}'.format(len(df[df.minimum_nights > sup_minimum_nights]),sup_minimum_nights))
print('Que representa {:.2f}% do dataset total'.format(len(df[df.minimum_nights > sup_minimum_nights])/df.shape[0]*100))
```

#### `price`

Iremos utilizar do mesmo racional utilizado na variável `minimim_nights` para fazer a limpeza da variável `price`.

A partir do limite superior encontrado para a variável `price`, iremos considerar o valor máximo de 287.5, que representa pouco menos de 7% do dataset total.

Por mim, considerando os trade off dos dados retirados, iremos desconsiderar cerca de 19% de nosso dataset total.

```
# price
fig, ax = plt.subplots()
df.price.plot(kind='box', vert = False, figsize=(12,3), ax=ax)
ax.set_title('\nBoxplot `price`\n')
plt.show()

# ver quantidade de valores acima do limite superior encontrado anteriormente para price
print('\nO dataset possui {} entradas'.format(df.shape[0]))
print('Para a variável `price` conta com {} entradas acima de {}'.format(len(df[df.price > sup_price]),sup_price))
print('Que representa {:.2f}% do dataset total'.format(len(df[df.price > sup_price])/df.shape[0]*100))
```

## Preparação dos dados

### Limpeza dos valores outliers e ausentes

Após a identificar a presença de valores ausentes e outliers em nosso dataset, iremos seguir com a limpeza dos dados. Está etapa do projeto é de grande influência no resultado final da análise. Conforme dito anteriormente, foram tomadas certas premissas. Tais considerações devem ser feitas com propriedade e com a conciência do trade-off incidente na decorrência do trabalho.

Tendo isso em vista, as variáveis alvo `price` e `minimum_nights` serão ajustadas para que se adeque aos limites superiores encontrados anteriormente. 

Junto à isso, serão realizados as tratativas já mencionadas acima para os elementos ausentes.

Após o tratamento das informações, os histogramas e boxplots serão plotados novamente. Com isso, é possível obter uma comparação da distribuição das entradas originais e tratadas das variáveis.

```
# fazer uma copia do dataframe
df_clean = df.copy()

# tratar variáveis `price` e `minimum_nights`.
df_clean = df_clean[df_clean.price < sup_price]
df_clean = df_clean[df_clean.minimum_nights < sup_minimum_nights]

# tratar dados ausentes da variável `reviews_per_month`.
df_clean.reviews_per_month.fillna(0, inplace=True)

# remover `neighbourhood_group` e `license`, pois estão vazias.
df_clean.drop(columns = ['neighbourhood_group', 'license', 'last_review'], axis=0, inplace = True)
```

## Análise exploratória

### Correlação entre as variáveis

Um artifício utilizado para análise de dados em dataset é a matriz de correlação.

A matriz de correlação, se trata de uma tabela que mostra o coeficientes de correlação existentes entre as diferentes variáveis. Ou seja, a partir dela, é possível verificar a interdependência das variáveis analisadas.

Com este artifício matemático somos capazes de explicitar as correlações entre as variáveis. Por isso, a matriz de correlação é considerada uma ótima ferramenta para resumir dataset extensos, a fim de identificar e visualizar padrões existentes.

Tendo isso em vista, primeiramente, será plotado uma matriz de correlação e, em seguida, um heatmap; para que seja possível visualizar as relações entre as variáveis de maneira mais intuitiva.

A partir desses plots, podem ser observados os seguintes pontos:

- A variável `number_of_reviews` tem alta correlação com `reviews_per_month`, o que é esperado, visto que quanto mais reviews um imóvel, maior tende a ser sua média mensal. Sendo assim, por terem forte correlação entre si, pode se considerar a retirada de uma das variáveis de nossa análise, tornando o DataFrame mais enxuto.

- Há uma correlação elevada entre `calculated_host_listings_count` e `availability_365`. Uma possível motivo para esta correlação, é devido a hipótese de que, anfitriões que possuem maior quantidade de imóveis, utilizam seus imóveis apenas para aluguel, deixando-os disponíveis para hospedes na maior parte do ano. Já para anfitriões com imóveis únicos, estes alugam suas próprias casas em que moram, deixando-as disponíveis durante um menor período de tempo.

```
# criar uma matriz de correlação
corrMatrix = df_clean[['price','minimum_nights','number_of_reviews','reviews_per_month','calculated_host_listings_count',
                       'availability_365']].corr()

# mostrar a matriz de correlação
corrMatrix
```

```
# plotar um heatmap a partir das correlações
sn.heatmap(corrMatrix, cmap="YlGnBu", fmt='.2f', annot=True)
plt.show()
```

#### Região de maior valor aquisitivo

Para identificar a localidade de Dublin com a maior média de preços, foi preciso utilizar a função `.groupby`. Está função é utilizada para agrupar valores de uma variável, tomando como referência valores agrupados de outra variável desejada.

No caso tomou-se como base o agrupamento da variável `price` a partir dos valores únicos da variável `neighbourhood`.

Junto a isso, foi utilizado o `.mean()` para encontrar a média de preços para cada localidade da cidade de Dublin.


```
# plot gráfico de barras
media_preco_neighbourhood = df_clean.groupby(['neighbourhood']).price.mean()
entradas_neighbourhood = df_clean.neighbourhood.unique()
precos_neighbourhood = pd.DataFrame({'preco_medio':media_preco_neighbourhood}, index=entradas_neighbourhood)

precos_neighbourhood = precos_neighbourhood.preco_medio.sort_values(ascending = False)

print('Médias de preços em cada região:\n{}\n'.format(precos_neighbourhood))

precos_neighbourhood.sort_values(ascending = True, inplace=True)

fig, ax = plt.subplots()
precos_neighbourhood.plot(kind='barh',figsize=((8,5)), ax=ax,)
ax.set_title('\nMédias de preços em cada região\n')
plt.tight_layout()
```

Foi realizada uma análise introdutória sobre a estruturação e criação de um projeto de análise de dados. O intuito do projeto foi apresentar artifícios estatísticos, gráficos e raciocínios necessários na construção de um trabalho envolvendo a análise de um dataset.

Tendo isso em vista, foi possível tirar alguns insight da base de dados, como: 
  - há a presença de outliers no dataset utilizado, sendo necessário o tratamento das informações; 
  - proporção disponível de cada tipo e imóvel;
  - correlação entre os valores das diferentes variáveis;
  - regiões maior caras para se alugar um imóvel na cidade de Dublin, Irlanda. 

Por fim, lembra-se que este dataset é uma versão resumida, ideal para uma abordagem inicial. Recomenda-se que seja usado, em uma próxima análise exploratória, o conjunto de dados completos, para uma análise mais sofisticada.
